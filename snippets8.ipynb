```python
# gcd
def divisors(a):
    res = [x for x in range(1,a+1) if a % x == 0]
    return res

def gcd(a,b):
    if b > a:
        c = a
        a = b
        b = c
    an = a
    bn = b
    rn = 1
    while rn > 0:
        rn = an % bn
        an = bn
        bn = rn
    return an

def gcd_iter(u, v):
    while v:
        u, v = v, u % v
    return abs(u)

def gcd_bin(u, v):
    u, v = abs(u), abs(v) # u >= 0, v >= 0
    if u < v:
        u, v = v, u # u >= v >= 0
    if v == 0:
        return u
    # u >= v > 0
    k = 1
    while u & 1 == 0 and v & 1 == 0: # u, v - even
        u >>= 1; v >>= 1
        k <<= 1
    t = -v if u & 1 else u
    while t:
        while t & 1 == 0:
            t >>= 1
        if t > 0:
            u = t
        else:
            v = -t
        t = u - v
    return u * k

def redfrac(num, denom): # reduce fraction
    ndgcd = gcd_bin(num, denom)
    return [ num / ndgcd, denom / ndgcd ]

if __name__=='__main__':
    #print gcd(24,16)
    print redfrac(234, 766)
```

```python
# restaurant visits
import random

def qual(nvis,nrk,nprob):
    res = []
    hiq = 0
    for ivis in range(nvis):
        if ivis < nprob:
            thisrk = random.randint(1,nrk)
            res.append(thisrk)
            if thisrk > hiq:
                hiq = thisrk
        else:
            res.append(hiq)
    return res
            
if __name__=='__main__':
    random.seed()
    totr = 0.
    for n in range(1000):
        totr += sum(qual(4,5,2))/4.
    print totr / 1000.
```
